```python
import pandas as pd
```

```python
df=pd.read_csv(r'../input/playground-series-s3e15/data.csv')
```

```python
df.describe()
```

```python
df = pd.get_dummies(df, dtype = 'int')
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=10)
df_filled = imputer.fit_transform(df)
df_filled = pd.DataFrame(df_filled, columns=df.columns)
```

```python
df_filled.head()
```

```python
submission = df_filled[['id', 'x_e_out [-]']]
submission.head()
```

```python
sample_submission = pd.read_csv(r'../input/playground-series-s3e15/sample_submission.csv')

# filter out all the id values from submission that are not in sample_submission
submission = submission[submission['id'].isin(sample_submission['id'])]
submission['id'] = submission['id'].astype(int)
```

```python
submission.to_csv('submission.csv', index=False)
```
